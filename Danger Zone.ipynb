The following code sets up the data frame, and the axes object, as well as plotting the ship positions. 

```python
import cartopy.crs as ccrs                   # import projections
import cartopy.feature as cf                 # import features
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('DangerZone.xlsx')

irma_lats = data['Average Latitude'].tolist()
irma_longs = data['Average Longitude'].tolist()
latstd=data['Latitude StD'].tolist()
longstd=data['Longitude StD'].tolist()


central_lon, central_lat = -40, 18
extent = [-50, -90, 15, 50]
axs = plt.axes(projection=ccrs.PlateCarree())
axs.stock_img()
axs.set_extent(extent)
axs.gridlines()
axs.coastlines(resolution='50m')
#setting up list for minimum and maximum latitude and longitude values
min_lasts=[]
max_lasts=[]
min_longs=[]
max_longs=[]
#create maximum values by adding std and minimum values by substracting std
position=0    #hold the values
for i in irma_lats:                              #loop through the latitudes
    min_lasts.append(i-latstd[position])
    max_lasts.append(i+latstd[position])
    position+=1        #to increament the value over and over again
position=0
for i in irma_longs:
    min_longs.append(i-longstd[position])
    max_longs.append(i+longstd[position])
    position+=1
#add the ship positions
axs.plot(-85,17,'ro', label="ship 1") 
axs.plot(-78,23,'ko', label="ship 2")
axs.plot(-68,18.5,'yo', label="ship 3")
axs.plot(-72,20,'bo', label="ship 4")
axs.plot(-73,17.70,'mo', label="ship 5")
axs.plot(-80,23,'go', label="ship 6")
fig_size=[20,15]                      #change figure size to make the map clearer
plt.rcParams['figure.figsize']=fig_size

#ploting min and mx path of the hurricane
axs.plot(min_longs,min_lasts,label='Minimum Path')
axs.plot(max_longs,max_lasts,label='Maximum Path')
axs.legend()
plt.show()
```
